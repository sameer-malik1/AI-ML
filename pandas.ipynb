```python
# apply(series & dataframe)

points = {
    '1st point':[(3,4),(4,5),(5,6),(4,4)],
    '2nd point':[(3,4),(4,3),(5,9),(5,5)]
}

p = pd.DataFrame(points)
p

def euclidean(p1,p2):
  return ((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)**0.5


p['distance'] = p.apply(lambda row: euclidean(row['1st point'],row['2nd point']),axis=1)
p
```

# **Groupby Object**

```python
import pandas as pd
import numpy as np
movies = pd.read_csv('/content/imdb-top-1000.csv')
movies.head(2)
```

```python
genres = movies.groupby("Genre")
```

```python
genres
```

```python
# applying builtin aggregation functions on groupby objects such as sum,mean,min,max
genres.min()
```

```python
# find out the top 3 genre which earns high profit at boxoffice
genres["Gross"].sum().sort_values(ascending=False).head(3)
```

```python
# find out the genre with highest avg imdb rating
genres["IMDB_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most popularity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find num of movies done by each actor using group by
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# find out how many goups have been formed
len(movies.groupby('Genre'))
```

```python
# find how many rows/recored each group contains
movies.groupby("Genre").size()
```

```python
# find first or last movie of each group
genres = movies.groupby('Genre')
genres.last()
```

```python
# geting all the recored a group contains
genres.get_group("Family")
```

```python
# groups -> an attribute, returns a dict with all the index of records each group contains
genres.groups
```

```python
# describe -> used to describe the overall summary of each group
genres.describe()
```

```python
genres.sample() # randomly shows the movies of each group
```

```python
genres.nunique() # gives the uniques valus for each group
```

```python
# aggregate functions specific to each of the columns
movies.groupby('Genre').agg({
    'Runtime':['max','max'],
    'No_of_Votes':'sum',
    'IMDB_Rating':['mean','std']

})
```

```python
# aggregaring based on list operations for each column
movies.groupby('Genre').agg(['min','max','sum'])
```

```python
# apply -> split,apply,combine
# find the num of movies starts with A

# movies[movies['Series_Title'].str.startswith("A")]  // simple approach

# using apply method
def func(group):
  return group['Series_Title'].str.startswith('A').sum()

movies.groupby('Genre').apply(func)
```

```python
# import ipl dataset for practice
import pandas as pd
ipl = pd.read_csv('/content/deliveries.csv')
ipl
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max num of sixes
sixes = ipl[ipl['batsman_runs']==6] # first filters the dataset to get only sixes by each player

# make groups based on batsman, then select/focus the batsman column from each group and counts the non-null values in each group’s 'batsman' column.
sixes.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most 4's and 6's in the last 5 over
last_over = ipl[ipl['over']>15] # filtering to get the dataset of last 5 overs
last_over = last_over[(last_over['batsman_runs']==6) | (last_over['batsman_runs']==4)] # filtering to get only 4 and 6 run data in the last over
last_over.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find Virat's run record against all teams
temp_df = ipl[ipl['batsman']=='V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum()
```
