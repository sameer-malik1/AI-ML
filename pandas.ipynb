**Series**: One-dimensional labeled array (like a column in a spreadsheet).

# **Filtering a DataFrame**

```python
df.head(20)
```

```python
# find the movies list which earned profit greater than 600
mask = df["Profitability"] > 2200
filteredDf = df[mask]
filteredDf[['Movie',"Story","Genre"]] # extracted the specific columns from filtered data frame using fancy indexing

# same output in one line
df[df["Profitability"]>2200][['Movie','Story','Genre']]
```

```python
# How many movies get the Audience Score greater than 80
df[df["AudienceScore"]>90][["Movie"]].shape[0]
```

```python
# How many movies get the Audience Score greater than 80 and its genre is comedy
df[(df["AudienceScore"]>80) & (df["Genre"] == "Comedy")][["Movie"]].shape[0]
```

```python
# how many movies earned more than 20% profit on total investment
df[(df["Profitability"] >= df["Budget"]*1.30) & (df["Profitability"] <= df["Budget"]*1.40)][['Movie',"Budget","Profitability","Year"]]
```

```python
# how many movies which include Man in their names
df[df["Movie"].str.split(" ").apply(lambda x:"Man" in x )]
```
