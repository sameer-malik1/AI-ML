# **Linear  Regression**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/placement.csv')
df.head()
```

```python
# plotting this dataset using scatter plot
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('Candidate CGPA')
plt.ylabel('Package in LPA')
```

```python
# before applying machine learning algorithm, seprate the x and y columns
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2) # spliting the data 20% for testng and 80% for training
```

```python
# importing the LinearRegression class
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train,y_train) # fit method is used to train the model
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('Candidate CGPA')
plt.ylabel('Package in LPA')
```
