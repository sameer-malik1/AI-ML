# **Preprocessing**

```python
# checking null values in each column
movies.isnull().sum()
```

```python
# removing null values if exist
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.genres.iloc[0]  # it gives the string, so we have to change it to list by parsing using ast
```

```python
import ast # abstract syntax tree is used to parse the data to actual data structure

def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
convert(movies.genres.iloc[0]) # testing function
```

```python
# converting and simplifying actual column
movies.genres = movies.genres.apply(convert)
movies.keywords = movies.keywords.apply(convert)
movies.head()
```

```python
# write a function to parse top 3 casts
def convert_cast(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter += 1
    else:
      break
  return L
```

```python
movies.cast = movies.cast.apply(convert_cast)
movies.head()
```

```python
 # write a function to get just director form crew column
 def get_director(obj):
  dl = []
  for i in ast.literal_eval(obj):
    if i['job'] == "Director":
      dl.append(i['name'])
      break
  return dl
```

```python
# apply the get_director function on crew column
movies.crew = movies.crew.apply(get_director)
movies.crew
```

```python
movies.overview = movies.overview.apply(lambda x:x.split())
movies.overview[0]
```

```python
movies.head(1)
```

```python
# removing spaces of names in the list of each column
movies.genres = movies.genres.apply(lambda x:[i.replace(" ","") for i in x])
movies.cast = movies.cast.apply(lambda x:[i.replace(" ","") for i in x])
movies.keywords = movies.keywords.apply(lambda x:[i.replace(" ","") for i in x])
movies.crew = movies.crew.apply(lambda x:[i.replace(" ","") for i in x])
movies.head(2)
```

```python
# concatenating the columns and named it 'tags' and add it to movies dataframe
movies['tags'] = movies.overview+movies.genres+movies.keywords+movies.cast+movies.crew
movies.tags.iloc[0]
```

```python
# creating new dataframe by extracting the most relavent columns
new_df = movies[['movie_id','title','tags']]
new_df
```

```python
# joining the tags from list to str
new_df.tags = new_df.tags.apply(lambda x:" ".join(x))
new_df.tags[0]
```

```python
new_df.tags = new_df.tags.apply(lambda x:x.lower())
new_df.tags[0]
```

# **Stemming**

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df['tags'][0]
```

```python
# write a function for stemming the each tag
def stem(text):
  l = []
  for i in text.split():
    l.append(ps.stem(i))
  return " ".join(l)
```
